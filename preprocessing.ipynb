```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df
```

```python
## data cleaning

df=df.drop(columns=['RowNumber','CustomerId','Surname'])
df
```

```python
df.isnull().sum()
```

```python
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df
```

```python
ohe=OneHotEncoder()
encoded=ohe.fit_transform(df[['Geography']]).toarray()
encoded
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
dff=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(['Geography']))
dff
```

```python
df=df.drop('Geography',axis=1)
```

```python
df
```

```python
df=pd.concat([df,dff],axis=1)
```

```python
df
```

```python
#with  open('le.pkl','wb') as file:
   # pickle.dump(le,file)
filename='le.sav'
pickle.dump(le,open(filename,'wb'))
```

```python
#with open ('ohe.pkl','wb') as file:
    #pickle.dump(ohe,file)
filename='ohe.sav'
pickle.dump(ohe,open(filename,'wb'))
```

```python
x=df.drop('Exited',axis=1)
y=df['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
#with open ('scaler.pkl','wb') as file:
 #   pickle.dump(scaler,file)
filename='scaler.sav'
pickle.dump(scaler,open(filename,'wb'))
```

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
model=Sequential([
    Dense(64,input_shape=(x_train.shape[1],),activation='relu'),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
]
)
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
earlystopping=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
log_dir='log_fit'
tensorboard=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
model.fit(
    x_train,y_train,
    validation_data=(x_test,y_test),
    epochs=100,
    callbacks=[earlystopping,tensorboard]
)
```

```python
model.save('model.weights.h5')
```
